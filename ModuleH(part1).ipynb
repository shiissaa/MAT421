**22.1 ODE Initial Value Problem Statement**

A differential equation represents the relationship between a function and its derivatives. Differential equations are used to model systems in every engineering and science field. Finding the exact solution to a differential equation is very hard, unlike differentiation where analytic solutions are more easily computed. This is why numerical solutions are critical, as they are what actually enable differential equations useful to be in their respective field.

**22.2 Reduction of Order**

Because many numerical methods for solving initial value problems are specifically designed to solve first-order differential equations, we can reduce the order of higher order differential equations to the first order. **22.3 The Euler Method**

The Euler Method allows us to compute an approximation for the state S at S(t_j+1). It is the most intuitive way for solving initial value problems.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
%matplotlib inline

# Parameters
f = lambda t, s: np.exp(-t) # diff eq
h = 0.1 # Step size
t = np.arange(0, 1+h, h) # Grid
s0 = -1 # Initial condition/value

# Euler method
s = np.zeros(len(t))
s[0] = s0
for i in range(0, len(t) - 1):
  s[i+1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

Each dot represents an approximation of the exact value, with each dot being one approximation based on the previous dot. We can adjust the h value, or step size, to get a better approximation.
