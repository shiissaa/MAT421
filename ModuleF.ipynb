**20.2 Finite Difference Approximating Derivatives**

Finite difference approximations of the slope allow us to use values of the function in the neighbordhood of the point x = a to achieve the goal. There are various formulas used in different applications

Forward difference is to estimate the slope of the function at $x_j$ using the line that connects $(x_j,f(x_j))$ and the points using j+1.

Backward difference is to estimate the slope of the function at $x_j$ using the line that connects the points using j-1 and $(x_j,f(x_j))$.

Central difference is to estimate the slope of the function at $x_j$ using hte line that connects the points using j-1 and j+1.

O(h) can be used to describe the accuracy of the used finite difference approximations. O(h) represents really $h(a+(h))$, where a is a constant and e(h) is a function that goes to 0 as h approaches 0.

The following code will compute the derivatives numerically using the forward difference technique where f(x) = cos(x). It will display an offset between two curves, which is a result from the numerical error, whose maximum is printed after the graph (~0.05). 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

# Determine step size, grid, function
h = 0.1
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)

# Compute the vector of forward differences, grid, and solution
forward_diff = np.diff(y)/h
x_diff = x[:-1:]
exact_sol = -np.sin(x_diff)

# Plot
plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', \
         label = 'Finite difference approx.')
plt.plot(x_diff, exact_sol, \
         label = 'Exact solution')
plt.legend()
plt.show()

# Compute max error between the exact answer and numerical approximation
max_error = max(abs(exact_sol - forward_diff))
print(max_error)
```

**20.4 Numerical Differentiation with Noise**

Noise refers to the contamination of data that occurs as a result of innaccuracies in measurement devices or the data itself being slightly modified by factors outside the system of interest. When noise occurs, the value of the data is off by a small amount from what it would be had it been computed by a pure mathematical function. We can display this effect in Python.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

x = np.arange(0, 2*np.pi, 0.01)
o = 100
e = 0.05
y = np.cos(x)
y_noise = y + e*np.sin(o*x)

plt.figure(figsize = (12, 8))
plt.plot(x, y_noise, 'r-', \
         label = 'cos(x) + noise')
plt.plot(x, y, 'b-', \
         label = 'cos(x)')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```
