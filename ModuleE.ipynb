**3.2 Continuity and Differentiation**

*Limits and Continuity*

Limits are concerned with the behavior of a function near a particular input. With python's sympy, we can easily find the limit of a function as it approaches a value.

```python
from sympy import *
x = Symbol('x')
f = sin(x)/x
limit(f, x, oo)
```

*Derivatives*

The derivative of a function of a real variable measures the sensitivity to change of the function value with respect to another variable. This can occur with one or more variables.

```python
# Finding the derivative of a function with one variable
x = Symbol('x')
f = x**3 - 9
fp = f.diff(x)
fp
```

```python
# Finding the derivative of a function with more than one variable (with respect to that variable)
x, y, z = symbols('x y z')
f = 5*x*y**2 - 3*y*z + 2*x*z**3
fp = f.diff(y)
fp
```

*Taylor's Theorem*

Taylor's Theorem gives an approximation of a differentiable function around a given point by a polynomial. The following shows the Taylor series approximations of the sine function.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order', 'Fifth Order', 'Seventh Order']

plt.figure(figsize = (10, 8))
for n, label in zip(range(4), labels):
  y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
  plt.plot(x, y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Apprxomations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

It is also possible to compute an nth order Taylor series approxmation for a function at a certain point.

```python
# Computing the seventh order Taylor series approximation for sin(x) at x = pi/2
x = np.pi/2
y = 0

for n in range(4):
  y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)

print(y)
```

**3.3 Unconstrained Optimization**

Unconstrained optimization involves finding the maximum or minimum of a differentiable function of several variables over a nice set. The goal is to derive optimality conditions for unconstrained continuous optimization problems.

*Necessary and Sufficient Conditions of Local Minimizers*

Global minimizer:  Let f : R^d → R. The point x∗ ∈ R^d is a
global minimizer of f over R^d if
f(x) ≥ f(x∗), ∀x ∈ R^d .

Local minimizer: Let f : R^d → R. The point x∗ ∈ R^d is a
local minimizer of f over R^d if there is δ>0 such that
f(x) ≥ f(x∗), ∀x ∈ Bδ(x∗)\ {x∗}

Descent direction: Let f : R^d → R. A vector v is a descent
direction for f at x0 if there is α∗ > 0 such that
f (x0 + αv) <f (x0), ∀α ∈ (0, α∗).

First-order necessary condition: Let f : R^d → R be continuously differentiable on R^d. If x0 is a local minimizer, then ∇f (x0)=0.

Second-order necessary condition: Let f : R^d → R be twice continuously differentiable on R^d . If x0 is a local minimizer, then Hf(x0) is positive semi-definite.

*Convexity and Global Minimizers*

Convex set: A set D ⊆ R^d is convex if for all x,y ∈ D and
all α ∈ [0,1], (1− α)x+αy ∈ D.

Convex function: A function f : R^d → R is convex if, for all x,y ∈ R^d and all α ∈ [0,1], f ((1− α)x+αy) ≤ (1− α)f(x)+ αf(y).

First-order convexity condition: Let f : R^d → R be continuously differentiable. Then f is convex if and only if for all x,y ∈ R^d f(y)≥f(x)+∇f(x)^T(y− x).
